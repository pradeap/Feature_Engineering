#### Target Guided Ordinal  Encoding

1. Ordering the labels according to the target
2. Replace the labels by the joint probability of being 1 or 0

```python
import pandas as pd
df = pd.read_csv('titanic.csv', usecols = ['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing', inplace = True)
df.head(20)
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head(10)
```

```python
df.Cabin.unique()
```

```python
df.groupby(['Cabin'])['Survived'].mean()
# probability of the survival rate depending on the cabin 
```

```python
df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels = df.groupby(['Cabin'])['Survived'].mean().sort_values().index
```

```python
ordinal_labels
```

```python
ordinal_labels2 = {k:i for i,k in enumerate(ordinal_labels,0)} ## check again
```

```python
ordinal_labels2 
```

```python
df['Cabin_ordinal_labels'] = df['Cabin'].map(ordinal_labels2)
```

```python
df.head()
```

####  Mean Encoding

```python
mean_ordinal = df.groupby(['Cabin'])['Survived'].mean().to_dict()
```

```python
df['mean_ordinal'] = df['Cabin'].map(mean_ordinal)
df.head()
```

##### Probability Ration Encoding


1. Probability of survived based on cabin -- categorical feature
2. probability of not survived 
3. ratio of survived/died
4. dictionary to map cabin with probability
5. replace with the categorical feature


```python
df = pd.read_csv('titanic.csv', usecols = ['Cabin','Survived'])
df.head()
```

```python
df['Cabin'].fillna('Missing',inplace = True)
df.head(20)
```

```python
df['Cabin'].unique()
```

```python
df['Cabin'] = df['Cabin'].astype(str).str[0]
df.head()
```

```python
df['Cabin'].unique()
```

```python
prob_df = df.groupby(['Cabin'])['Survived'].mean()
```

```python
prob_df
```

```python
prob_df = pd.DataFrame(prob_df)
prob_df
```

```python
prob_df['Died'] = 1 - prob_df['Survived']
```

```python
prob_df.head()
```

```python
prob_df['Probability_ratio'] = prob_df['Survived']/prob_df['Died']
```

```python
prob_df.head()
```

```python
probabilty_encoded = prob_df['Probability_ratio'].to_dict()
```

```python
df['Cabin_encoded'] = df['Cabin'].map(probabilty_encoded)
```
